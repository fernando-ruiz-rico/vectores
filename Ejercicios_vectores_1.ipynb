# Concepto de Vectores

## ¿Qué es un vector?
Un vector es una lista de números que tienen tanto una magnitud como una dirección. En términos sencillos, un vector es como una flecha que apunta hacia una dirección específica y tiene una longitud (o magnitud).

Los vectores son muy útiles en aplicaciones de machine learning porque nos permiten representar datos de manera estructurada. Por ejemplo, si queremos representar diferentes características de una casa (como su tamaño, número de habitaciones y su antigüedad), podemos usar un vector para almacenar esta información en una sola lista. Esto facilita realizar cálculos y procesar los datos en el ordenador.

### Ejemplo 1: Representación de características de una casa usando un vector

Supongamos que estamos creando un modelo para predecir el precio de una casa. Para ello, necesitamos representar varias características de la casa, como:
- Tamaño en metros cuadrados
- Número de habitaciones
- Antiguedad de la casa (en años)

Podemos almacenar estas características en un vector. Esto nos permite tener toda la información de la casa en una lista, lo cual es más ordenado y eficiente que manejar cada dato por separado. Así, podemos enviar esta lista directamente al modelo de machine learning para que realice cálculos con ella.

En Python, esto se vería así:

```python
import numpy as np

caracteristicas_casa = np.array([120, 3, 10])  # Tamaño en m. cuadrados, número de habitaciones, antiguedad en años
print("Vector de características de la casa:", caracteristicas_casa)
```

### Ejemplo 2: Puntuaciones de estudiantes en un examen representadas como un vector
Cada componente de un vector también puede representar diferentes valores relacionados entre sí. Por ejemplo, si tenemos tres estudiantes y queremos almacenar sus puntuaciones en un examen, podemos usar un vector donde cada posición representa la puntuación de un estudiante.

Vector:
- Estudiante 1: 7 puntos
- Estudiante 2: 8 puntos
- Estudiante 3: 9 puntos

Al usar un vector, podemos analizar rápidamente las puntuaciones de los estudiantes y compararlas entre sí.

```python
puntuaciones = np.array([7, 8, 9])
print("Puntuaciones de los estudiantes:", puntuaciones)
```

### Ejemplo 3: Representación de la velocidad y dirección del viento
Un vector también puede representar la velocidad y dirección del viento. Si tenemos un vector $[5, 2]$, esto puede significar que el viento sopla a 5 m/s en la dirección horizontal (x) y a 2 m/s en la dirección vertical (y).

En el mundo real, los meteorólogos usan vectores para representar el viento en mapas. Esto les permite ver fácilmente la dirección del viento en diferentes áreas y su intensidad.

```python
import matplotlib.pyplot as plt

# Creamos primero el vector del viento
viento = np.array([5, 2])

# Representamos el vector en el gráfico
plt.quiver(0, 0, viento[0], viento[1], angles='xy', scale_units='xy', scale=1, color='blue')
plt.xlim(-1, 6)
plt.ylim(-1, 3)
plt.xlabel("Dirección horizontal (m/s)")
plt.ylabel("Dirección vertical (m/s)")
plt.title("Representación del vector de viento en 2D")
plt.grid()
plt.show()
```

### Ejemplo 4: Representación de plantas

Imaginemos una planta con tres características: longitud de pétalo, ancho de pétalo y longitud de sépalo. Podemos representar estas características en un vector.

Al representar las características de una planta en un vector, podemos comparar fácilmente las plantas entre sí y utilizar esta información en modelos de machine learning que clasifiquen o analicen diferentes tipos de plantas.

```python
caracteristicas_planta = np.array([3.5, 1.4, 5.1])
print("Características de la planta:", caracteristicas_planta)
```
